```python
import matplotlib.pyplot as plt
import numpy as np

# Create data points for x and y
x = np.linspace(-10, 10, 100)  # 100 points between -10 and 10
y = x**2  # Calculate y values

# Plot the data
plt.plot(x, y)

# Customize the plot
plt.title("$y = x^2$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

# Display the plot
plt.show()
```
